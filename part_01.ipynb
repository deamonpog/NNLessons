## Single Neuron

A Neuron with 3 inputs [x1, x2, x3], weights [w1, w2, w3] and a output value




```python
# Neuron of 3 inputs 
inputs = [1, 2, 3]
weights = [0.2, 0.8, -0.5]
bias = 2.0
output = (inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2] * weights[2] + bias)
output
```

#### Using Numpy
np.dot function gives vector dot product for two 1D arrays

```python
# Using numpy
output = np.dot(weights, inputs) + bias
output
```

## A Layer of Neurons

A layer of 3 neurons with 3 inputs [x1, x2, x3] 



```python
inputs = [1, 2, 3]
weights = [[0.2, 0.8, -0.5],
           [0.1, 0.4, 0.2],
           [-0.4, 0.4, -0.3]]
biases = [2.0, 3.0, -0.5]

output = [  (inputs[0] * weights[0][0] + inputs[1] * weights[0][1] + inputs[2] * weights[0][2] + biases[0]),
            (inputs[0] * weights[1][0] + inputs[1] * weights[1][1] + inputs[2] * weights[1][2] + biases[1]),
            (inputs[0] * weights[2][0] + inputs[1] * weights[2][1] + inputs[2] * weights[2][2] + biases[2]) ]
output
```

#### Using Numpy
np.dot function gives vector dot product of each row of first param with the second param when the two params are matrix, vector

```python
# Using Numpy
output = np.dot(weights, inputs) + biases
output
```

## A layer of Neurons with batches of inputs

Same as a layer of neurons but with an input matrix which contains a input in its each row

```python
inputs = [[1, 2, 3],
          [4, 5, 6]]
weights = [[0.2, 0.8, -0.5],
           [0.1, 0.4, 0.2],
           [-0.4, 0.4, -0.3]]
biases = [2.0, 3.0, -0.5]

output1 = [ (inputs[0][0] * weights[0][0] + inputs[0][1] * weights[0][1] + inputs[0][2] * weights[0][2] + biases[0]),
            (inputs[0][0] * weights[1][0] + inputs[0][1] * weights[1][1] + inputs[0][2] * weights[1][2] + biases[1]),
            (inputs[0][0] * weights[2][0] + inputs[0][1] * weights[2][1] + inputs[0][2] * weights[2][2] + biases[2]) ]

output2 = [ (inputs[1][0] * weights[0][0] + inputs[1][1] * weights[0][1] + inputs[1][2] * weights[0][2] + biases[0]),
            (inputs[1][0] * weights[1][0] + inputs[1][1] * weights[1][1] + inputs[1][2] * weights[1][2] + biases[1]),
            (inputs[1][0] * weights[2][0] + inputs[1][1] * weights[2][1] + inputs[1][2] * weights[2][2] + biases[2]) ]

outputs = [output1, output2]
outputs
```
