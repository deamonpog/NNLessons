```python
import numpy as np
```

# Adding more layers



```python
inputs = [[1, 2, 3],
          [4, 5, 6]]

weights_0 = [[0.2, 0.8, -0.5],
             [0.1, 0.4, 0.2],
             [-0.4, 0.4, -0.3],
             [0.5, -0.5, 0.5]]
biases_0 = [0.2, 0.8, -0.5, 1.0]

weights_1 = [[0.5, -0.5, 0.5, 1.0],
             [0.2, 0.8, -0.5, -1.0],
             [-0.4, 0.4, -0.3, 0.5]]
biases_1 = [-0.4, 0.4, -0.3]

weights_2 = [[0.1, 0.4, 0.2],
             [-0.4, 0.4, -0.3],
             [0.5, -0.5, 0.5]]
biases_2 = [0.1, 0.4, 0.2]
```

```python
output_0 = np.matmul( inputs, np.array(weights_0).T ) +  biases_0
output_0
```

```python
output_1 = np.matmul( output_0, np.array(weights_1).T ) +  biases_1
output_1
```

```python
output_2 = np.matmul( output_1, np.array(weights_2).T ) +  biases_2
output_2
```
